### DFS 내 답

```python
n, m = map(int, input().split())
mold = [list(map(int, input().split())) for _ in range(n)]
```

```python
visited = [[0]*m for _ in range(n)]
icecream = 0
dx = [1, -1, 0, 0]
dy = [0, 0, -1, 1]

def searching(x,y):
    visited[x][y] = 1
    # 좌표 =상하좌우 탐색
    for i in range(4):
        nx, ny = x + dx[i], y + dy[i]
        # 주어진 공간 내에서
        if 0 <= nx < n and 0 <= ny < m:
            # 방문하지 않았거나 몰드가 있다면
            if visited[nx][ny] == 0 and mold[nx][ny] == 0:
                # 방문!
                visited[nx][ny] = 1
                # 해당 좌표에서 다시 상하좌우 탐색
                searching(nx, ny)
    return 1            

for x in range(n):
    for y in range(m):
        # 위치가 방문했거나 틀에 막혔다면 continue
        if visited[x][y] == 1 or mold[x][y] == 1:
            continue

        else:
            search = searching(x, y)
            if search != 0:
                icecream += 1

print(icecream)
```

### DFS 모법답

```python
def searching(x, y):
    # 범위 밖이거나, 이미 방문했거나, 틀(1)이면 탐색 중지
    if x < 0 or x >= n or y < 0 or y >= m:
        return False
    if visited[x][y] == 1 or mold[x][y] == 1:
        return False

    # 현재 노드 방문 처리
    visited[x][y] = 1
    
    # 상하좌우 재귀 호출 (리턴값은 필요 없음, 방문 처리만 하면 됨)
    # 여기서 return을 쓰지 않고 그냥 4방향을 다 찔러보는 게 핵심!
    searching(x - 1, y)
    searching(x, y - 1)
    searching(x + 1, y)
    searching(x, y + 1)
    
    return True  # 탐색을 성공적으로 마쳤으니 True 반환

# 메인 실행부
icecream = 0
for i in range(n):
    for j in range(m):
        # searching이 True를 반환하면(새로운 덩어리를 찾으면) 카운트 +1
        if searching(i, j) == True:
            icecream += 1

print(icecream)
```

```python
# N, M을 공백을 기준으로 구분하여 입력 받기
n, m = map(int, input().split())

# 2차원 리스트의 맵 정보 입력 받기
graph = []
for i in range(n):
    graph.append(list(map(int, input())))

# DFS로 특정한 노드를 방문한 뒤에 연결된 모든 노드들도 방문
def dfs(x, y):
    # 주어진 범위를 벗어나는 경우에는 즉시 종료
    if x <= -1 or x >= n or y <= -1 or y >= m:
        return False
    # 현재 노드를 아직 방문하지 않았다면
    if graph[x][y] == 0:
        # 해당 노드 방문 처리
        graph[x][y] = 1
        # 상, 하, 좌, 우의 위치들도 모두 재귀적으로 호출
        dfs(x - 1, y)
        dfs(x, y - 1)
        dfs(x + 1, y)
        dfs(x, y + 1)
        return True
    return False

# 모든 노드(위치)에 대하여 음료수 채우기
result = 0
for i in range(n):
    for j in range(m):
        # 현재 위치에서 DFS 수행
        if dfs(i, j) == True:
            result += 1

print(result) # 정답 출력
```

### BFS 내 답 - 툴람

```python
n, m = map(int, input().split())
miro = [list(input().strip()) for _ in range(n)]
```

```python
for i in range(n):
    for j in range(m):
        miro[i][j] = int(miro[i][j])
```

```python
cost = []
print(miro)
def search(x, y, cnt):
    if x < 0 or x >= n or y < 0 or y >= m:
        return False

    cnt+= 1
    if x == n-1 and y == m-1:
        cost.append(cnt)
        return
        
    if miro[x][y] == 1:
        miro[x][y] = 0
        search(x-1, y, cnt)
        search(x+1, y, cnt)
        search(x, y-1, cnt)
        search(x, y+1, cnt)


search(0, 0, 0)
cost.sort()
print(cost[0])
```

### BFS 모범답

```python
n, m = map(int, input().split())
miro = [list(input().strip()) for _ in range(n)]
```

```python
for i in range(n):
    for j in range(m):
        miro[i][j] = int(miro[i][j])
```

```python
from collections import deque

dx = [1, -1, 0, 0]
dy = [0, 0, -1, 1]

def search(x, y):
    queue = deque()
    queue.append([x,y])

    while queue:
        x, y = queue.popleft()

        for i in range(4):    
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            if miro[nx][ny] == 0:
                continue

            if miro[nx][ny] == 1:
                miro[nx][ny] = miro[x][y] + 1
                queue.append([nx, ny])

    return miro[n-1][m-1]
      

print(search(0,0))

```
