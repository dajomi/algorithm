```python
start = [1,1]

# D [1,0] U [-1, 0] R [0,1] L [0,-1]
for i in move:
    if (i == 'R') & (start[1] < n):
        start[1] += 1
    elif (i == 'L') & (start[1] > 1):
        start[1] -= 1
    elif (i == 'U') & (start[0] > 1):
        start[0] -= 1
    elif (i == 'D') & (start[0] < n):
        start[0] += 1

print(start[0], start[1])
```

```python
# N 입력받기
n = int(input())
x, y = 1, 1
plans = input().split()

# L, R, U, D에 따른 이동 방향
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
move_types = ['L', 'R', 'U', 'D']

# 이동 계획을 하나씩 확인
for plan in plans:
    # 이동 후 좌표 구하기
    for i in range(len(move_types)):
        if plan == move_types[i]:
            nx = x + dx[i]
            ny = y + dy[i]
    # 공간을 벗어나는 경우 무시
    if nx < 1 or ny < 1 or nx > n or ny > n:
        continue
    # 이동 수행
    x, y = nx, ny

print(x, y)
```

```python
# 현재 나이트의 위치 입력받기
input_data = input()
row = int(input_data[1])
column = int(ord(input_data[0])) - int(ord('a')) + 1

# 나이트가 이동할 수 있는 8가지 방향 정의
steps = [(-2, -1), (-1, -2), (1, -2), (2, -1), (2, 1), (1, 2), (-1, 2), (-2, 1)]

# 8가지 방향에 대하여 각 위치로 이동이 가능한지 확인
result = 0
for step in steps:
    # 이동하고자 하는 위치 확인
    next_row = row + step[0]
    next_column = column + step[1]
    # 해당 위치로 이동이 가능하다면 카운트 증가
    if next_row >= 1 and next_row <= 8 and next_column >= 1 and next_column <= 8:
        result += 1

print(result)
```

```python
n,m = map(int, input().split())
loc = list(map(int, input().split()))

game_map = []
for i in range(n):
    game_map.append(list(map(int, input().split())))
```

```python
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]
dir = [0, 1, 2, 3] # dir[i-1]
cur_loc = [loc[0], loc[1]]
cur_dir = loc[2]
went = [[0] * m for _ in range(n)]
went[cur_loc[0]][cur_loc[1]] = 1
cnt = 1

def is_no_went(temp_loc):
    if went[temp_loc[0]][temp_loc[1]] == 1: # 갔던 곳이면 False
        return False
    else: # 안 갔던 곳이면 True
        return True
        
def is_no_sea(temp_loc):
    if game_map[temp_loc[0]][temp_loc[1]] == 1:
        return False
    else:
        return True

def is_all_went(temp_loc):
    cnt = 0
    for i in range(4):
        cur_loc = [temp_loc[0] + dx[i], temp_loc[1] + dy[i]]
        if (is_no_went(cur_loc) == False) or (is_no_sea(cur_loc) == False):
            cnt += 1
    if cnt == 4:
        return True
```

```python
while True:

    # 모든 칸을 갔거나, 모든 칸이 바다인 경우 
    if is_all_went(cur_loc):
        print("모든 칸을 갔거나 모든 칸이 바다예요")

        # 한 칸 뒤로
        print("뒷 칸으로 가볼게요")
        cur_loc = [cur_loc[0] + dx[(cur_dir + 2) % 4], cur_loc[1] + dy[(cur_dir + 2) % 4]]
        
        # 근데 뒷칸이 바다인 경우 -> 중지
        if is_no_sea(cur_loc) == False:
            print("근데 여기가 바다예요. 종료할게요.", cur_dir, cur_loc)
            break

    else:
        # 왼쪽으로 돌기
        cur_dir = dir[cur_dir-1]
    
        # 안 간 곳이거나 바다가 아니면 옮기기
        temp_loc = [cur_loc[0] + dx[cur_dir], cur_loc[1] + dy[cur_dir]]
        if is_no_went(temp_loc) and is_no_sea(temp_loc):
            cur_loc = temp_loc
            went[cur_loc[0]][cur_loc[1]] += 1 # 갔다는 표시
            cnt += 1
            print("바다가 아니고, 안 간 곳이어서 한 칸 왔어요")
            
    
print(cnt)
```

```python
# 입력 받기
n, m = map(int, input().split())
# 방문 위치를 저장하기 위한 맵을 생성하여 0으로 초기화
d = [[0] * m for _ in range(n)]

# 캐릭터의 현재 위치, 방향 입력 받기
x, y, direction = map(int, input().split())
d[x][y] = 1 # 현재 위치 방문 처리

# 전체 맵 정보 입력 받기
array = []
for i in range(n):
    array.append(list(map(int, input().split())))

# 북, 동, 남, 서 방향 정의 (문제의 0,1,2,3 순서에 맞춤)
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

# 왼쪽으로 회전하는 함수
def turn_left():
    global direction
    direction -= 1
    if direction == -1:
        direction = 3

# 시뮬레이션 시작
count = 1 # 방문한 칸의 수 (시작점 포함하니까 1부터)
turn_time = 0 # 회전 횟수 체크

while True:
    # 1. 왼쪽으로 회전
    turn_left()
    nx = x + dx[direction]
    ny = y + dy[direction]
    
    # 2. 회전한 이후 정면에 가보지 않은 칸이 존재하고, 바다가 아니라면 이동
    # (d[nx][ny] == 0 : 안 가본 곳, array[nx][ny] == 0 : 육지)
    if d[nx][ny] == 0 and array[nx][ny] == 0:
        d[nx][ny] = 1 # 방문 체크
        x = nx
        y = ny
        count += 1
        turn_time = 0 # 이동했으니 회전 횟수 초기화
        continue
    
    # 3. 회전한 이후 정면에 가보지 않은 칸이 없거나 바다인 경우
    else:
        turn_time += 1 # 회전 횟수만 늘림
    
    # 4. 네 방향 모두 갈 수 없는 경우 (4번 돌았을 때)
    if turn_time == 4:
        nx = x - dx[direction] # 뒤로 갈 좌표 계산
        ny = y - dy[direction]
        
        # 뒤로 갈 수 있다면 이동 (육지라면)
        if array[nx][ny] == 0:
            x = nx
            y = ny
        # 뒤가 바다로 막혀있는 경우
        else:
            break
        turn_time = 0 # 이동했으니(혹은 멈췄으니) 초기화

print(count)
```
