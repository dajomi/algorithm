### 큰 수의 법칙

#### 내 답안

```python
n, m, k = map(int, input().split()) # 5, 8, 3
data = list(map(int, input().split()))

data = sorted(data)

first = data[n-1]
second = data[n-2]

count = m // (k+1) * k + (m % (k+1)) 

sum = 0
sum += first * count
sum += second * (m = count)

print(sum)
```

#### 모범 답안

```python
n, m = map(int, input().split()) # 2 4

max = 0
for i in range(n):
    nums = list(map(int, input().split()))
    nums.sort()
    if nums[0] > max:
        max = nums[0]

print(max)
```

### 숫자 카드 게임

#### 내 답안

```python
n, m = map(int, input().split()) # 2 4

max_val = 0
for i in range(n):
    nums = list(map(int, input().split()))
    min_value = min(nums)                         # 쓸데없는 정렬 버림
    if min_value > max_val:
        max_val = min_value

print(max_val)
```

#### 모범답안

```python
n, m = map(int, input().split()) # 2 4

result = 0
for i in range(n):
    data = list(map(int, input().split()))
    min_value = min(data)
    result = max(result, min_value)

print(result)
```

### 1이 될 때까지

내 답안

```python
n, k = map(int, input().split())
sum = 0

while n >= k:
    if (n % k == 0) :
        n = n // k
        sum += 1
    else:
        tmp = n % k
        sum += tmp
        n -= tmp
sum += (n-1)
print(sum)
```

#### 모범 답안

```python
n, k = map(int, input().split())

result = 0

while True:
    target = (n // k) * k            # 17 5 -> target = 15
    result += (n - target)           # result = 2
    n = target                       # n = 15

    if n < k:
        break
        
    n >= k:
        n = n // k                   # n = 3
        result += 1                  # result = 3

        
```
