```python
arrs = [list(map(int, input().split())) for _ in range(9)]

maxi = [0,0,0] # 값, 행, 열
for i in range(9):
    now_max = max(arrs[i])
    if maxi[0] < now_max:
        maxi = [now_max, i, arrs[i].index(now_max)]

print(maxi[0])
print(maxi[1]+1, maxi[2]+1)
```

```python
n = int(input())
papers = [list(map(int,input().split())) for _ in range(n)]
big = [[0]*100 for _ in range(100)]

for paper in papers:
    col = paper[0]
    row = paper[1]
    for i in range(row, row+10):
        big[i][col:(col+10)] = [1]*10
        
col_sum = [sum(big[i]) for i in range(100)]
print(sum(col_sum))
```
