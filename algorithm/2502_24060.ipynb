```python
n = int(input().strip())
nodes = {}
for i in range(n):
    node = list(input().split())
    nodes[node[0]] = [node[1], node[2]]
```

```python
def pre(root, node):
    print(root, end = '')
    if node[0] != '.':
        pre(node[0], nodes[node[0]])
    if node[1] != '.':
        pre(node[1], nodes[node[1]])
        

def mid(root, node):
    if node[0] != '.':
        mid(node[0], nodes[node[0]])
    print(root, end = '')
    if node[1] != '.':
        mid(node[1], nodes[node[1]])


def post(root, node):
    if node[0] != '.':
        post(node[0], nodes[node[0]])
    if node[1] != '.':
        post(node[1], nodes[node[1]])
    print(root, end = '')
```

```python
root = 'A'
pre(root, nodes[root])
print()
mid(root, nodes[root])
print()
post(root, nodes[root])
```

```python
n, k = map(int, input().split())
array = list(map(int, input().split()))

```

```python
cnt = 0
```

```python
def merge_sort(array, p, r):
    if (p < r):
        q = (p + r) // 2 
        merge_sort(array, p, q)      # 전반부 정렬
        merge_sort(array, q + 1, r)  # 후반부 정렬
        merge(array, p, q, r)        # 병합

def merge(A, p, q, r):
    tmp = []
    i = p
    j = q + 1
    while (i <= q and j <= r):
        if (A[i] <= A[j]):
            tmp.append(A[i])
            i += 1
            global cnt
            cnt += 1
        else:
            tmp.append(A[j])
            j += 1
            cnt += 1

        if cnt == k:
            pass
    
        while (i <= q):  # 왼쪽 배열 부분이 남은 경우
            tmp.append(A[i])
            i += 1
        while (j <= r):  # 오른쪽 배열 부분이 남은 경우
            tmp.append(A[j])
            j += 1
            
    return tmp
```
