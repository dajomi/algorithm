```python
h, m = map(int, input().split())

hour = h
min = m - 45

# 시간이 바뀔 때
if min < 0:
  min = 60 + min
  hour = hour -1
  # 바뀐 시간이 23시 일 때
  if hour == -1:
    hour = 23

print(hour, min)
```

```python
h, m = map(int, input().split())
time = int(input().strip())

hour = h
min = m + time

if min > 59:
  plus_hour = min // 60
  hour += plus_hour
  min = min % 60
  if hour > 23:
    hour = hour - 24

print(hour, min)
```

```python
n = int(input().strip())
arr = list(map(int, input().split()))
x = int(input().strip())

cnt = 0
arr.sort()

start = 0
end = n-1

while start < end:
  if (arr[start] + arr[end] == x):
    cnt += 1
    start += 1
  elif (arr[start] + arr[end] < x):
    start += 1
  elif (arr[start] + arr[end] > x):
    end -= 1

print(cnt)
```
